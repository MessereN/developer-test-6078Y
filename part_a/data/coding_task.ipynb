```python

expression_data = pd.read_csv('icb_gene_expression.csv')
clinical_data = pd.read_csv('icb_clinical_data.csv')
expression_data
```

```python
expr_sample_ids = set(expression_data.columns[1:])


# Get patient IDs from clinical dataset
clin_sample_ids = set(clinical_data["patientid"])

# Find columns in expression but not in clinical
missing_in_clinical = clin_sample_ids - expr_sample_ids

print(missing_in_clinical)
```

```python
# Count missing values in each clinical dataset column
missing_clinical_cols = clinical_data.isna().sum()

# Calculate percentage missing for each sample (row) in clinical data
missing_per_sample = clinical_data.isna().mean(axis=1) * 100

# Flag samples with more than 10% missing fields
flagged_samples = clinical_data.loc[missing_per_sample > 10, "patientid"]
print("\nMissing values per clinical column:")
print(missing_clinical_cols)
print("\nSamples with >10% missing clinical fields:")
print(flagged_samples.tolist())
```

```python
# Calculate mean expression for each gene (skip first column which is gene ID)
expression_data["mean_expression"] = expression_data.iloc[:, 1:].mean(axis=1)

# Sort by mean expression (descending) and select top 5
top5_genes = expression_data.sort_values(by="mean_expression", ascending=False).head(5)

# Output
top5_genes[["Unnamed: 0", "mean_expression"]].to_csv("top5_genes.csv")
print(top5_genes[["Unnamed: 0", "mean_expression"]])
```

```python
import pandas as pd

# Load datasets
expr_df = pd.read_csv("icb_gene_expression.csv")
clin_df = pd.read_csv("icb_clinical_data.csv")

# Get the list of common sample IDs
expr_sample_ids = set(expression_data.columns[1:])
clin_sample_ids = set(clinical_data["patientid"])
common_ids = sorted(expr_sample_ids & clin_sample_ids)

# Subset expression data to keep only common IDs
expr_common = expr_df[["Unnamed: 0"] + common_ids]

# Transpose expression data so samples become rows and gene IDs are columns
expr_common_t = expr_common.set_index("Unnamed: 0").T
expr_common_t.index.name = "patientid"

# Merge with clinical data on patientid
merged_df = pd.merge(expr_common_t.reset_index(), clin_df, on="patientid", how="inner")

print(merged_df.shape)
merged_df.to_csv("merged_data.csv")
merged_df.head()
```
